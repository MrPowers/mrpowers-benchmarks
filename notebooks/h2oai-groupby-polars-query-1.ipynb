```python
import polars as pl
```

```python
dtypes={
    "id1": pl.Utf8,
    "id2": pl.Utf8,
    "id3": pl.Utf8,
    "id4": pl.Int32,
    "id5": pl.Int32,
    "id6": pl.Int32,
    "v1": pl.Int32,
    "v2": pl.Int32,
    "v3": pl.Float64,
}
```

```python
medium = "../tmp/N_1e8_K_1e2_single.csv"
large = "../tmp/N_1e9_K_1e2_single.csv"
```

```python
medium_df = pl.scan_csv(medium, dtypes=dtypes)
```

```python
large_df = pl.scan_csv(large, dtypes=dtypes)
```

```python
%%time
medium_df.groupby("id1").agg(pl.sum("v1")).collect()
```

```python
%%time
large_df.groupby("id1").agg(pl.sum("v1")).collect()
```
